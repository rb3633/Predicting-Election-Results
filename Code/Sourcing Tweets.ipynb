```python
import snscrape.modules.twitter as sntwitter 
import pandas as pd
import itertools
```

```python
counties = pd.read_csv('us-county-boundaries.csv', delimiter=';', skiprows=0, low_memory=False)

counties
```

```python
county_list = counties.copy()[['Geo Point', 'NAME']]

county_list
```

```python
coords = list(county_list['Geo Point'])

coords
```

```python
coords = [coord+', 10km' for coord in coords]

coords
```

```python
# Save counties coordinates to a file
import json

# with open('county_coordinates.txt', 'w') as f:
#     f.write(json.dumps(coords))
```

```python
#Now read the file back into a Python list object
with open('county_coordinates.txt', 'r') as f:
    coords = json.loads(f.read())

# coords
```

```python
import re
```

### Obama 2012

```python
#CURRENT TWEET SOURCE CODE

# Creating list to append tweet data to
tweets_list = []
checker = 0
# Using TwitterSearchScraper to scrape data and append tweets to list
for coord in coords:
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper('obama geocode:"{}" since:2012-10-05 until:2012-11-05'.format(coord)).get_items()):
        if None:
            #print('No tweet found ',i)
            continue
        if i>100:
            #print('15 done')
            break
        geopoint=coord.replace(', 10km','')
        location=str(counties.loc[counties['Geo Point']==geopoint,'NAME']+', '+counties.loc[counties['Geo Point']==geopoint,'STUSAB'])
        location=re.match('\d+ (.+)\\ndtype.+',location)
        loc=location.groups()
        tweets_list.extend([[tweet.user.username, tweet.date, tweet.rawContent, tweet.likeCount, loc[0], tweet.user.location]])
    #checker+=1
    
    #if checker>3:
    #   break

        
# Creating a dataframe from the tweets list above
obama_12 = pd.DataFrame(tweets_list, columns=['Username','Datetime', 'Content', 'Likes', 'GeoLocation', 'ProfileLocation'])

obama_12
```

```python
obama_12.to_csv('obama_12_geotagged.csv')
```
